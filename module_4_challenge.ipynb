## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
data_file_df['subtotal'] = data_file_df['unit_price'] * data_file_df['qty']
print(data_file_df)

#data_file_df['subtotal'].astype(int)
#print(data_file_df['subtotal'])
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
data_file_df['total_weight'] = data_file_df['unit_weight'] * data_file_df['qty']

data_file_df['shipping_price'] = data_file_df['total_weight'].apply(lambda x: 7 if x > 50 else 10)

print(data_file_df)

#data_file_df['total_weight'].astype(int)
#data_file_df['shipping_price'].astype(int)
#print(data_file_df['total_weight'])
#print(data_file_df['shipping_price'])
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
data_file_df['line_price'] = (data_file_df['subtotal'] + data_file_df['shipping_price'] * data_file_df['total_weight']) * 1.0925

print(data_file_df)

#data_file_df['line_price'].astype(int)
#print(data_file_df['line_price'])
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
data_file_df['shipping_cost'] = data_file_df["shipping_price"] * data_file_df["total_weight"]
data_file_df['line_cost'] = data_file_df["unit_cost"] * data_file_df["qty"] + data_file_df["shipping_cost"]

print(data_file_df)

#data_file_df['shipping_cost'].astype(int)
#data_file_df['line_cost'].astype(int)
#print(data_file_df['shipping_cost'])
#print(data_file_df['line_cost'])
```

```python
# Create a column for the profit of
# each line using line cost and line price
data_file_df['profit'] = data_file_df['line_price'] - data_file_df['line_cost']
print(data_file_df)

#data_file_df['profit'].astype(int)
#print(data_file_df['profit'])
```

Use the new columns with confirmed values to find the following information.

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

summary_df = data_file_df.groupby('client_id').agg({
    'qty': 'sum',
    'shipping_cost': 'sum',
    'line_price': 'sum',
    'profit': 'sum',
}).reset_index()

top_clients_id = [33615, 66037, 46820, 38378, 24741]

summary_df = summary_df.sort_values('profit', ascending = False)

top_five_summary = summary_df[summary_df['client_id'].isin(top_clients_id)]

print(top_five_summary)
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

reformatted_summary_df = top_five_summary.rename(columns = {'client_id': 'Client ID', 'qty': 'Total Units Purchased', 'shipping_cost': 'Total Shipping', 'line_price': 'Total Revenue', 'profit': 'Total Profit', '': ''})

reformatted_summary_df['Total Units Purchased'] = reformatted_summary_df['Total Units Purchased'].astype(int)
reformatted_summary_df['Total Shipping'] = (reformatted_summary_df['Total Shipping']).astype(int)
reformatted_summary_df['Total Revenue'] = (reformatted_summary_df['Total Revenue']).astype(int)
reformatted_summary_df['Total Profit'] = (reformatted_summary_df['Total Profit']).astype(int)

reformatted_summary_df['Total Units Purchased'] = reformatted_summary_df['Total Units Purchased'].map('{:,}'.format)
reformatted_summary_df['Total Shipping'] = (reformatted_summary_df['Total Shipping']/1000000).map('${:,}M'.format)
reformatted_summary_df['Total Revenue'] = (reformatted_summary_df['Total Revenue']/1000000).map('${:,}M'.format)
reformatted_summary_df['Total Profit'] = (reformatted_summary_df['Total Profit']/1000000).map('${:,}M'.format)

print(reformatted_summary_df)

```

```python
# Sort the updated data by "Total Profit" form highest to lowest

custom_order = [144, 319, 621, 402, 258]  # Define a custom order of indices
df_reordered = reformatted_summary_df.reindex(custom_order)

print(df_reordered)
```
